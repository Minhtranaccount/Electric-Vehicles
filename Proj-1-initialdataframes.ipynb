```python
#Import Dependencies
import pandas as pd
```

```python
#read EV data CSV file

file="IEA-EV-data.csv"

data=pd.read_csv(file)
data.head()
```

```python
#Clean & check data
df=pd.DataFrame(data)
df=df.dropna(how='any')
df.dtypes
```

```python
#Filter for historical sales only, exclude 'World'
filtered_df=df.loc[(df['category']=='Historical') & (df['parameter'] == 'EV sales') & (df["region"] != "World")]
filtered_df
```

```python
#Create sales specific df for future usage
sales_df = pd.DataFrame(filtered_df.groupby(['year','region']).sum())
sales_df
```

```python
sales_df.to_csv('EV-Sales-Data.csv')
```

```python
#Filter for historic charging stations only, exclude 'World'
filtered_df2=df.loc[(df['category']=='Historical') & (df['parameter'] == 'EV charging points')& (df["region"] != "World")]
filtered_df2
```

```python
#Create charging specific df for future usage
charging_df= pd.DataFrame(filtered_df2.groupby(['year','region']).sum())
charging_df
```

```python
charging_df.to_csv('EV-Charging_Data.csv')
```
